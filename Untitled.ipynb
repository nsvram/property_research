```python
for lst in my_data_lst:
    print(lst)
    zf = zipfile.ZipFile(lst)
    try:
        zf.extractall(zf.filename.split(".")[0])
    except KeyError:
        print ('ERROR: Did not find %s in zip file')
    else:
        print("Loaded data into " + zf.filename.split(".")[0])
```

```python
for i in my_data_lst :
    if i.endswith(".DAT"):
        print(i)
```

```python
A_str = ""
B_str = ""
C_str = ""
D_str = ""
Z_str = ""

for i in my_data_lst:
    with open(i) as f:
        lis=[line.split() for line in f]
        for i in lis:
            if ("".join(i)).split(";")[0] == 'A':
                A_str =  A_str + ",".join(("".join(i)).split(";")) + "|" 
            if ("".join(i)).split(";")[0] == 'B':
                B_str = B_str + ",".join(("".join(i)).split(";")) + "|"
            if ("".join(i)).split(";")[0] == 'C':
                C_str = C_str + ",".join(("".join(i)).split(";")) + "|"
            if ("".join(i)).split(";")[0] == 'D':
                D_str = D_str + ",".join(("".join(i)).split(";")) + "|"
            if ("".join(i)).split(";")[0] == 'Z':
                Z_str = Z_str + ",".join(("".join(i)).split(";")) + "|"
```
